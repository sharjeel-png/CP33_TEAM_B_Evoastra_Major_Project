**ðŸš€ STEP 6 â€” Clean & Standardize Column Names**

```python
# Clean column names to make them easier to work with in Python.
# Steps:
# 1. Convert all column names to lowercase â†’ avoids case issues
# 2. Replace spaces/dashes with underscores â†’ avoids syntax errors
df.columns = (
    df.columns
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)

# Check the new column names
df.head()
```

**ðŸš€ STEP 7 â€” Handle Missing Values**

```python
# 'income' contains missing values.
# We fill missing values using the MEDIAN instead of mean:
# - Mean is affected by outliers
# - Median gives a more stable central value

df['income'].fillna(df['income'].median(), inplace=True)

```

```python
# Remove rows where 'dt_customer' is missing.
# Date is essential for customer analysis, so we drop rows missing this value.
# df.dropna(subset=['dt_customer'], inplace=True)
```

```python
# Check again to ensure missing values are fixed
df.isnull().sum()
```

**ðŸš€ STEP 8 â€” Removing Duplicate Records**

```python
# Remove duplicate rows â€” duplicates can distort analysis.
df.drop_duplicates(inplace=True)
```

**ðŸš€ STEP 9 â€” Converting Date Column to Datetime Format**

```python
# Convert the dt_customer column into proper datetime format.
# errors='coerce' converts invalid dates into NaT (missing date).

df['dt_customer'] = pd.to_datetime(df['dt_customer'], errors='coerce')

# Check the result
df['dt_customer'].head()
```

**ðŸš€ STEP 10 â€” Remove Invalid Dates (NaT)**
     - These appeared AFTER datetime conversion

```python

# Count how many invalid dates were created
print("Invalid dates before removing:", df['dt_customer'].isnull().sum())

# Remove rows where dt_customer could not be converted (NaT values)
df = df.dropna(subset=['dt_customer'])

# Confirm removal
print("Invalid dates after removing:", df['dt_customer'].isnull().sum())
```

```python
# Checking the result again
df['dt_customer'].head()
```

**ðŸš€ STEP 11 â€” Converting Numeric Columns to Correct Types**

```python
# Columns that should be integers
numeric_columns = ['kidhome', 'teenhome', 'income', 'recency']

# Converting them to integer type
df[numeric_columns] = df[numeric_columns].astype(int)

df.dtypes
```

**ðŸš€ STEP 12 â€” Feature Engineering (Adding useful New Columns)**

```python
# Adding Customer Age

df['age'] = 2024 - df['year_birth']
```

```python
# Add Total Customer Spending

spending_columns = [
    'mntwines',
    'mntfruits',
    'mntmeatproducts',
    'mntfishproducts',
    'mntsweetproducts',
    'mntgoldprods'
]

df['total_spent'] = df[spending_columns].sum(axis=1)
```

**ðŸš€ STEP 13 â€” Final Dataset Validation**

```python
print(df.info())       # Check data types
```

```python
print(df.isnull().sum())  # Confirm no missing critical values
```

```python
# Export cleaned dataset to a new CSV file
df.to_csv("cleaned_customer_personality.csv", index=False)

print("Cleaned dataset saved successfully!")
```

1. What are missing values and how do you handle them?

   Missing values are blanks or null entries in a dataset where information is unavailable.
  

    They can be handled by:
- Removing rows or columns with too many missing values
- Imputing using mean, median, or mode
- Forward/Backward filling for time-series data
- Using models to predict missing values

---
2. How do you treat duplicate records?
Duplicate rows can distort analysis. They are treated by:
df.drop_duplicates()
This removes repeated rows and ensures data integrity.

---
3. Difference between dropna() and fillna() in Pandas?
dropna(): Removes rows or columns that contain missing values.
fillna(): Fills missing values with a specific value (mean, median, mode, or constant).
---
4. What is outlier treatment and why is it important?
Outliers are extreme values that deviate significantly from other observations.
Treatment methods:
- IQR method
- Z-score method
- Capping or flooring
Outlier treatment is important because outliers can skew averages, distort models, and produce misleading insights.

5. Explain the process of standardizing data.
Standardizing data means converting values into a consistent format. This includes:
- Making text lowercase
- Removing extra spaces
- Converting categories to uniform names (e.g., MALE â†’ male)
- Scaling numeric features (optional)
Helps ensure consistency and avoids errors in analysis.
